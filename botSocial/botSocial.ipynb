```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import random
import getpass
class startInstagram:
        def __init__(self,bot,usuario,senha,target=None,t=2,viewuser=False):
            self.bot = bot
            if target is None:
                bot.openweb('https://www.instagram.com')
            else:
                bot.openweb('https://www.instagram.com/'+target)
            time.sleep(t)
            if viewuser is True:
                print('login com:',usuario)
            if senha is not None:
                print('Autenticado com a senha.')
            pusuario ='//input[@name="username"]'
            psenha = '//input[@name="password"]'
            bot.login(usuario,senha,pusuario,psenha)
            time.sleep(2+t)
            
        def comentar(self,target,texto,texto2,tmin=20,tmax=180):
            bot = self.bot
            for i in range(0,len(texto2)):
                bot.openweb(target)
                time.sleep(1.5)
                bot.click('//textarea[@aria-label="Adicione um comentário..." ]')
                print(i+1,'comentarios')
                print('  ',texto2[i])
                #bot.click('//textarea[@aria-label="Adicione um comentário..." ]')
                time.sleep(random.randint(2,5))
                bot.inserirtexto('//textarea[@aria-label="Adicione um comentário..." ]',random.choice(texto)+' '+texto2[i])
                bot.click('//*[@id="react-root"]/section/main/div/div[1]/article/div[3]/section[3]/div/form/button')
                time.sleep(random.randint(tmin,tmax))

        def comentar_hashtag(self,hashtag,texto,rolar,seguir=True):
            bot = self.bot
            bot.openweb('https://www.instagram.com/explore/tags/'+hashtag+'/')
            time.sleep(2)
            bot.rolarbarra(rolar)
            hrefs = bot.coletarinfo('//a[@href]',attribute='href')
            print(hashtag + ' fotos ' + str(len(hrefs)))
            count=1
            for pic_href in hrefs:
                print(count,'comentários')
                count+=1
                bot.openweb(pic_href)
                time.sleep(1)
                try:
                    if seguir is True:
                        bot.click('//button[@class="sqdOP yWX7d    y3zKF     "]')
                    bot.inserirtexto('//textarea[@aria-label="Adicione um comentário..."]',random.choice(texto) )
                    time.sleep(random.randint(2,5))
                    time.sleep(2)
                except Exception as e:
                    print(e)
                    time.sleep(5)

        def seguidores(self,link,target):
            bot = self.bot
            bot.openweb(link+target+'/followers/')
            time.sleep(1)
            bot.click('//a[@href="/'+target+'/followers/"]')
            try:            
                input('Pressione ENTER para continuar.')
            except SyntaxError:
                pass
            hrefs = bot.coletarinfo('//a[@class="FPmhX notranslate  _0imsa "]')
            print(len(hrefs))
            return hrefs
```
